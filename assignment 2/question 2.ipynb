```python
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.base import MIMEBase
from email import encoders

def send_email(subject, to_address, cc_address, bcc_address, message_body, attachment_paths):
    # Replace these placeholders with your *email* credentials
    smtp_server = 'mmtp.iitk.ac.in'
    smtp_port = 25
    sender_email = 'adityarm22@iitk.ac.in'
    sender_password = 'Choti@bachchi74'

    # Create a connection to the SMTP server
    server = smtplib.SMTP(smtp_server, smtp_port)
    server.starttls()

    # Login to the email account
    server.login(sender_email, sender_password)

    # Create a MIMEMultipart object
    msg = MIMEMultipart()

    # Add sender, recipient, cc, and bcc information
    msg['From'] = sender_email
    msg['To'] = to_address
    msg['Cc'] = cc_address
    msg['Bcc'] = bcc_address

    # Add subject and body to the email
    msg['Subject'] = subject
    msg.attach(MIMEText(message_body, 'plain'))

    # Attach files
    for attachment_path in attachment_paths:
        attachment = open(attachment_path, 'rb')
        part = MIMEBase('application', 'octet-stream')
        part.set_payload((attachment).read())
        encoders.encode_base64(part)
        part.add_header('Content-Disposition', "attachment; filename= %s" % attachment_path)
        msg.attach(part)

    # Convert the message to a string and send it
    server.sendmail(sender_email, to_address, msg.as_string())

    # Quit the server
    server.quit()


# Example usage
subject = 'Test Email'
to_address = 'shobhits21@iitk.ac.in'
cc_address = 'adityarm22@iitk.ac.in'
bcc_address = ''
message_body = 'Hemlo friends chai peelo'
attachment_paths = []

send_email(subject, to_address, cc_address, bcc_address, message_body, attachment_paths)
```
